```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import random

# Creating dataset
x,y=np.loadtxt("nac/pc3_pc4.dat",comments=["#",'@'],unpack=True,dtype=float)
xedges, yedges = np.linspace(-4,4,20), np.linspace(-2,2,20)
hist, xedges, yedges = np.histogram2d(x, y, (xedges, yedges))
xidx = np.clip(np.digitize(x, xedges), 0, hist.shape[0]-1)
yidx = np.clip(np.digitize(y, yedges), 0, hist.shape[1]-1)

c = hist[xidx, yidx]
norm=np.max(c)

plt.figure(figsize=(8,6))
plt.scatter(x, y, c=-np.log(c/norm))
#plt.scatter(x, y, c=c/norm)
plt.xlabel("PC3",fontsize=15)
plt.ylabel("PC4",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

cbar = plt.colorbar()
cbar.set_label("Free Energy (in the unit of $k_bT$)",fontsize=20)
for t in cbar.ax.get_yticklabels():
     t.set_fontsize(26)
plt.tight_layout()
plt.savefig("nac_pc3_pc4.png",dpi=300)
plt.show()
```
